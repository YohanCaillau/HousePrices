# <center> Preparation of the data, turn all features numerical </center>

## Load data

```python
from pandas import read_csv
import pandas as pd
train_full = read_csv('../HousePrices/seloger-paris-page1-to-200_raw.csv', index_col=0, sep="\t")
```

## Select only the categorical features

```python
train_full
```

```python
train_full.dtypes
```

## Split data

Let's do 80-20%:

```python
from numpy import random
SEED = 42
random.seed(SEED)
```

```python
from sklearn.model_selection import train_test_split
VAL_SIZE = 0.2
train, val = train_test_split(train_full, test_size=VAL_SIZE)
```

```python
train_object = train.select_dtypes(include='object')
train_object = train_object.fillna('None')
val_object = val.select_dtypes(include='object')
val_object = val_object.fillna('None')
```

```python
#pip install category_encoders
```

```python
import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder
ohe = ce.OneHotEncoder(handle_unknown='ignore', use_cat_names=True)

train_ohe = ohe.fit_transform(train_object)
train_ohe.head()
```

```python
val_ohe = ohe.transform(val_object)
val_ohe.head()
```

## Same treatment for the numerical features

```python
train_number = train.select_dtypes(exclude='object')
train_number = train_number.fillna(0)
val_number = val.select_dtypes(exclude='object')
val_number = val_number.fillna(0)
```

Concatener the numerical and categorical features into one dataframe trainfull

```python
import pandas as pd
train = pd.concat([train_number, train_ohe], axis=1)
train.head()
```

```python
val = pd.concat([val_number, val_ohe], axis=1)
val.head()
```

```python
print("Train shape: " + str(train.shape))
print("Val shape: " + str(val.shape))
```

## Save data

```python
train.to_csv("train_ohe_AZ.csv")
val.to_csv("val_ohe_AZ.csv")
```
